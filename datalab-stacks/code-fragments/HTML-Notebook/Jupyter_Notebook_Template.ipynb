Im nächsten Schritt wollen wir nur die PLZ behalten, welche dem Kanton Bern zugeordnet sind:

```python
# Import Excel file from BFS
df_bfs_be = pd.read_excel('Import/XLSX/BFS_Gemeinde_PLZ_20211130.xlsx', usecols = "A,B,E");

# Drop duplicates
df_bfs_be = df_bfs_be.drop_duplicates(keep='first')

# Rename columns
df_bfs_be = df_bfs_be.rename(columns={"PLZZ": "PLZ6", "KTKZ": "Kanton"})
swisstopo_PLZ = swisstopo_PLZ.rename(columns={"PLZ": "PLZ4", "ZUSZIFF": "PLZ6"})

# Only select PLZ of canton BE
filter = df_bfs_be["Kanton"] == "BE"
df_bfs_be_short = df_bfs_be.where(filter).dropna()

# Left join the two tables
swisstopo_PLZ_comb = pd.merge(df_bfs_be_short, swisstopo_PLZ, on=['PLZ4', 'PLZ6'], how='left')

# Reorder pandas dataframe
swisstopo_PLZ_comb = swisstopo_PLZ_comb[["PLZ4","PLZ6","geometry"]];

# Convert pandas dataframe to geopandas dataframe
swisstopo_PLZ_comb = geopandas.GeoDataFrame(swisstopo_PLZ_comb)

# Quick and dirty plot of geopandas dataframe
swisstopo_PLZ_comb.plot()

# Plot and axis titles
plt.title("PLZ des Kanton Bern")
plt.xlabel("CSR LV95 (x)")
plt.ylabel("CSR LV95 (y)");
```
